# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
path=r"crime_data.csv"
df1=pd.read_csv(path)
df1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
df1_stand=scaler.fit_transform(df1.iloc[:,1:])
```

```python
plt.figure(figsize=(15,15))
dendrogram=sch.dendrogram(sch.linkage(df1_stand,method='single',metric='euclidean',))
```

```python
hc=AgglomerativeClustering(n_clusters=12,affinity='euclidean',linkage='single')
```

```python
hc.fit_predict(df1_stand)
```

```python
df1['cluster_id']=hc.labels_
```

```python
df1
```

```python
df1.cluster_id.value_counts()
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
df2=pd.read_csv(path)
df2
```

```python
wcss=[]
for i in range(1,24):
    kmean=KMeans(n_clusters=i,random_state=42)
    kmean.fit(df1_stand)
    wcss.append(kmean.inertia_)
plt.plot(range(1,24),wcss)
```

# For 12 th point there is constant decrease so we consider optimum no of clusters as 12

```python
kmean=KMeans(n_clusters=12,random_state=2)
```

```python
kmean.fit_predict(df1_stand)
```

```python
cluster_ids=kmean.labels_
```

```python
df2['cluster id']=cluster_ids
```

```python
df2
```

```python
df2.columns
```

```python
df2['cluster id'].value_counts()
```
